```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 14})
```

```python
x = 3  # True slope
a = np.arange(-1, 3, .1, dtype='float32')
a = a.reshape(-1, 1)
# unpack values, same as np.random.randn(a.shape[0], a.shape[1])
b = x*a + np.random.randn(*a.shape)
```

```python
U, S, VT = np.linalg.svd(a, full_matrices=False)
S = np.diag(S)
A_dagger = VT.T @ np.linalg.inv(S) @ U.T
x_tilde = A_dagger @ b
```

```python
plt.figure()

plt.scatter(a, b, marker='x', c='r', label='Noisy data')
plt.plot(a, x*a, c='k', linewidth=2, label='True line')
plt.plot(a, x_tilde*a, '--', c='b', linewidth=2, label='Linear Regression')

plt.xlabel('a')
plt.ylabel('b')
plt.grid(linestyle='--')
plt.legend()
plt.show()
```
